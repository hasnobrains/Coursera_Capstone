```python
import folium 
import matplotlib.cm as cm
import matplotlib.colors as colors
import numpy as np
```

```python
# create map

map_clusters = folium.Map(location=[55.685,37.578], zoom_start=13)


x = np.arange(df.shape[0])
ys = [i + x + (i*x)**2 for i in range(df.shape[0])]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

for lat, lon, poi, index in zip(df['latitude'], df['longitude'], df['name'], df.index):
    label = folium.Popup(str(poi), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        fill_opacity=0.7).add_to(map_clusters)
    
map_clusters
```
